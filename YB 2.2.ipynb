### 맷플롯립과 시본 개요



*   맷플롯립(Matplotlib)은 다른 파이썬 시각화 라이브러리에 많은 영향을 미쳤으며, 여전히 파이썬 세계에서 시각화를 위해 많이 쓰인다. 하지만 직관적이지 못한 개발 API로 인해 시각화 코딩에 익숙해지는 데 많은 시간이 필요하며, 차트의 축 이름이나 차트 타이틀, 범례(legend) 등의 부가적인 속성까지 코딩을 해주어야 하는 불편함이 있다. 또 맷플롯립의 기본 설정 환경에서는 현대적인 감각이 떨어지는 시각화 플롯 등의 문제점을 가지고 있다.
*   시본(Seaborn)은 맷플롯립보다 쉬운 구현, 수려한 시각화 그리고 판다스와의 편리한 연동을 특징으로 한다. 시본은 맷플롯립을 기반으로 하나 그보다 상대적으로 적은 양의 코딩으로도 보다 수려한 시각화 플롯을 제공한다. 또, 판다스의 칼럼명을 기반으로 자동으로 축 이름을 설정하는 등 편리한 연동 기능을 가지고 있다. 하지만 특정 요소들의 경우 맷플롯립 함수들을 그대로 사용하고 있기에 시본을 잘 활용하려면 반드시 맷플롯립을 어느 정도는 알고 있어야 한다.

### 맷플롯립(Matplotlib)

*   맷플롯립은 파이썬 시각화를 위한 기반 모듈인 pyplot을 제공하며 이를 통해 시각화를 구현한다. pyplot은 plot() 함수를 통해 선 그래프를 그릴 수 있다. 입력값으로 x 좌표에 해당하는 여러 데이터 값과 y좌표에 해당하는 여러 데이터 값을 입력받아서 x와 y값이 매핑되는 선 그래프를 그릴 수 있다. 또, title() 함수는 입력값으로 타이틀 문자열을 받아서 차트 타이틀을 설정할 수 있다. 마지막으로 show() 함수를 호출하여 축을 포함한 선 그래프를 화면에 나타낼 수 있다.
*   맷플롯립, Matplotlib은 이름에서 유추해 볼 수 있듯 유명한 공학용 수치 연산과 통계 프로그램인 MATLAB에서 시각화를 위해 개발한 라이브러리에서 유래한다. 그리고 pyplot 모듈은 MATLAB 사용자들이 시각화를 파이썬에 사용할 때보다 쉽게 적응할 수 있도록 설계된 모듈이기에 MATLAB 스타일의 인터페이스가 반영되었다. 그 때문에 파이썬 언어 사상을 반영한 다른 서드파티 라이브러리들보다는 상대적으로 단순함과 직관성 측면에서 떨어지는 부분이 있다.





```python
# pyplot 모듈을 (관용적으로) plt라는 이름으로 import.
import matplotlib.pyplot as plt

# x축 입력값 [1, 2, 3], y축 입력값 [2, 4, 6]으로 선 그래프 형성.
plt.plot([1, 2, 3], [2, 4, 6])

# 선그래프 타이틀을 Hello plot으로 설정.
plt.title("Hello plot")

# 선그래프를 출력
plt.show()
```

### pyplot의 두 가지 중요 요소 - Figure와 Axes 이해



*   Figure 객체는 그림을 그리기 위한 캔버스의 역할을 한다고 간주할 수 있다. 그림판의 크기를 조절한다든가, 플롯을 최종적으로 화면에 나타내는 역할을 수행한다. 하지만 실제로 그림을 그리는 역할은 수행하지 않는다.
*   Axes는 실제 그림을 그리는 메서드들을 가진다. 그래서 대부분의 시각화를 수행하는 메서드들은 Axes에서 호출된다. 또, X축, Y축, 타이틀, 범례 등의 속성을 설정하는 데에도 이용된다. Axes(축 Axis의 복수형)라는 이름이 주는 인상 때문에 단순히 X축이나 Y축에 관련된 설정 정도의 역할을 수행할 것이라는 선입견과는 다르게 시각화와 관련된 많은 작업들이 이 Axes 객체 기반에서 수행된다. 반면 Figure는 그림을 그리는 실제 작업에는 크게 연관되어 있지 않다.

앞 절에서의 간단한 직선 그래프를 그리는 예제를 Figure와 Axes 관점에서 설명하면,

1.   plt.plot([1, 2, 3], [2, 4, 6])은 기본으로 설정된 Axes에서 Axes.plot() 함수를 호출해 그림을 그린다. Axes.plot()은 다시 Line2D 객체를 호출해 선 그래프를 그린다.
2.   plt.title("Hello plot")에서 plt.title()은 내부적으로 다시 Axes.set_title() 함수를 호출해 타이틀을 설정한다.

### Figure와 Axis의 활용



*   Figure은 주로 전체 그림판의 크기를 조절하는 데에 사용된다. plt 모듈의 figure() 함수를 호출하면 Figure 객체가 반환된다. 이때 인자로 figsize=(가로크기, 세로크기)를 입력하면 해당 크기를 가지는 figure 객체를 반환한다.
*   plt.figure() 함수에 의해 반환된 Figure가 실제로 그림을 그리는 많은 역할을 하지 못하지만 그림판의 배경 색깔을 바꾸는 등의 설정을 할 수 있다. 그러나 plt.figure(facecolor='color')을 적용해도 전체 그림판의 색이 바뀌는 것은 아닌데, Axes가 차지하는 영역이 Figure가 차지하는 그림판 위에 적용되기 때문이다.







```python
# plt.figure()는 주로 figure의 크기를 조절하는 데 사용된다
# 아래는 figure 크기가 가로 10, 세로 4인 Figure 객체를 설정하고 반환한다

plt.figure(figsize = (10, 4))

plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Hello plot")
plt.show()
```

```python
figure = plt.figure(figsize = (10, 4))
print(type(figure))
```

```python
plt.figure(figsize = (4, 4), facecolor = 'yellow')  # figure() 함수의 인자로 facecollor='yellow'를 통해 그림판의 배경색을 노란색으로 변경한다
plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Hello plot")
plt.show()
```



*   plt의 axes() 함수는 현재 사용하는 Axes 객체를 반환한다. plt.axes()로 반환된 객체 AxesSubplot은 Axes의 실체화된 하위 클래스로 보면 된다. plt.axes()만 호출했는데도 축을 포함한 그림이 표출된 이유는 pyplot 모듈이 import되어 있는 상태에서 주피터 노트북은 별도의 plt.show()가 없더라도 Axes가 할당되어 있으면 자동으로 plt.show()와 같은 역할을 하는 함수를 호출하면서 그림이 표출되게 하기 때문이다.
*   pyplot에서 Figure와 Axes 객체를 함께 가져올 수 있는데 이는 plt.subplots()을 이용한다. plt의 subplots() 함수는 인자로 여러 개의 Axes를 설정할 수 있지만, 여기서는 디폴트 인자만 적용해 단 한 개의 Axes만 가져올 수도 있다.
*   Axes 객체의 plot(), set_title()을 이용해서 앞의 pyplot의 plot()과 title()을 이용해 그린 선 그래프와 동일한 결과를 얻을 수 있다.






```python
ax = plt.axes()
print(type(ax))
```

```python
fig, ax = plt.subplots(figsize = (4, 4))  ## pyplot 모듈의 plot()이나 title()이 아닌 Axes 객체를 바로 이용해 선 그래프를 그린다

# Axes 객체의 plot()을 이용하여 선 그래프를 그린다
ax.plot([1, 2, 3], [2, 4, 6])

# Axes 객체의 set_title()을 이용해 선 그래프 타이틀을 Hello plot으로 설정한다
ax.set_title('Hello plot')  ## 여기서는 "" 아닌 ''
plt.show()
```

### 여러 개의 plot을 가지는 subplot들을 생성하기



*   plt.subplots() 함수는 주로 여러 개의 subplot을 생성하는 데 활용한다. 즉, 하나의 Figure 상에서 여러 개의 그래프를 그릴 수 있게 만들어주는데, 이 때 개별 subplot은 하나의 Axes를 가지게 된다. 즉 plt.subplots()은 여러 개의 그래프를 그릴 수 있게 하는 1개의 Figure와 여러 개의 Axes를 생성하고 반환한다.
*   plt.subplots()의 주요 인자로 nrows와 ncols, 그리고 figsize가 사용된다. nrow는 전체 subplot들의 배치를 2차원 행렬 형태로 표현할 때 행의 개수가 되며, ncols는 열의 개수가 된다. 그리고 figsize는 모든 subplot들을 포함한 전체 Figure의 크기를 설정한다. 여러 개의 Axes 객체는 튜플 형태 혹은 넘파이의 ndarray 형태로 받을 수 있다.



```python
# 아래는 두 개의 subplot, 즉 두 개의 Axes를 생성하되 열 방향으로 두 개를 배치한다
# 또한 전체 크기가 가로 6, 세로 3인 Figure을 생성한다

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))
```

### pyplot의 plot() 함수를 이용해 선 그래프 그리기



*   맷플롯립은 그래프/차트를 포함한 많은 시각화 객체와 함수를 제공한다. pyplot의 plot() 함수는 선 그래프를 그릴 때 활용된다. X 좌표 값, Y 좌표 값으로 파이썬 리스트, 넘파이 ndarray, 판다스의 DataFrame/Series 모두 적용 가능하다. 다만 입력되는 X 좌표 값과 Y 좌표 값의 개수는 모두 같은 크기를 가져야 한다. 또, plot() 함수는 color 인자를 이용해 선의 색깔을 변경할 수 있다. 하지만 선 그래프를 보다 다양하게 변화시키고 싶다면 더욱 다양한 인자값을 알고 있어야 한다.
*   엑셀과 태블로, 파워 BI와 같은 시각화 솔루션에서 시각화를 구현하는 것과 달리 맷플롯립과 시본 같은 라이브러리에서 API 기반의 시각화를 구현하는 방식은 큰 차이가 있다. API 기반의 시각화는 필연적으로 코드를 이용해 해당 함수들에 대한 인자 값이나 시각화를 위한 여러 가지 주변 값들을 설정해줘야 한다. 때문에 함수별로 인자값이 상이하거나 시각화를 위한 여러 가지 설정값이 복잡할 경우 시각화 구현에 많은 시간을 쏟기 쉽다. 맷플롯립이나 시본과 같은 API 기반의 시각화 솔루션은 이러한 태생적인 문제점을 당연한 것으로 받아들여야 하지만, 맷플롯립의 API가 시본에 비해 상대적으로 훨씬 다양하고 더 작은 단위로 쪼개져 있어 코딩하는 데 더 많은 시간이 소모될 수 있다. 예를 들어 맷플롯립의 경우, X축과 Y축의 명칭을 설정하거나 범례를 설정하는 작업을 모두 별도의 함수들을 사용해 구현해야 한다. 또, 데이터 분석에서 사용되는 여러 차트 등을 만들 때 여러 개의 함수들을 조합해서 만들어야 할 수도 있다. 때문에 맷플롯립보다는 시본을 활용해 데이터 분석을 위한 시각화를 수행하는 경우가 더 많다.



```python
import numpy as np
import pandas as pd

x_value = [1, 2, 3, 4]
y_value = [2, 4, 6, 8]
#x_value = np.array([1, 2, 3, 4])  # 넘파이 array로 테스트하려면 주석을 해제하고 실행
#y_value = np.array([2, 4, 6, 8])  # 넘파이 array로 테스트하려면 주석을 해제하고 실행

# 입력값으로 파이썬 리스트, numpy array 가능하다. x 좌표값과 y 좌표값은 모두 같은 크기를 가져야 한다
plt.plot(x_value, y_value)
plt.show()
```

```python
# color 인자를 green으로 설정해 선의 색깔을 녹색으로 변경한다
plt.plot(x_value, y_value, color='green')
```

```python
# API 기반으로 시각화를 구현할 때는 함수의 인자들에 대해서 알고 있어야 하는 부작용(?)이 있다
## 선의 색은 빨간색이고, 선의 형태는 대시형이고, 선의 두께는 2 그리고 각 좌표점별로 동그라미 형태의 마커를 크기 12로 설정해 선 그래프를 그린다

plt.plot(
    x_value, y_value, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=12
)
```

### 축 명칭 설정, 축의 눈금(틱)값 회전, 범례(legend) 설정하기



*   맷플롯립은 기본적으로 축의 명칭을 자동으로 설정하지 않는다. 축의 명칭을 설정하기를 원한다면 pyplot의 xlabel(), ylabel()과 같은 함수를 이용해야 한다.


```python
plt.plot(
    x_value, y_value, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=12
)

# x축과 y축의 이름을 텍스트로 할당하여 설정한다
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```



*   보통 축의 값이 문자열이고 많은 눈금값이 있을 경우 눈금 간의 값이 중복되면서 눈금값이 제대로 보이지 않을 때 축의 눈금값(틱값)을 회전할 수 있다. X축의 눈금값을 회전시키기 위해서는 xticks()를, Y축의 경우에는 yticks() 함수를 사용한다. 이때 보통 X축만 눈금값을 회전시킨다.



```python
# x 좌표 값을 0~99까지 100개로 설정한다
x_value = np.arange(0, 100)
# y 좌표 값은 개별 x좌표값을 2배로 하여 100개로 설정한다
y_value = 2 * x_value

plt.plot(x_value, y_value, color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')

# x축 눈금값을 90도로 회전한다
plt.xticks(rotation=90)

plt.show()
```

그런데 X좌표값을 0~99까지 100개로 설정했으나 0, 20, 40, 60, 80, 100 눈금 값들만 표시된다.


*   맷플롯립은 축 값이 문자열인 경우는 모든 값을 나타내주지만, 축 값이 숫자값이면 자동으로 최솟값과 최댓값에 기반해 축의 크기에 따라 눈금 간격을 만들어준다. X축의 눈금 값을 좀 더 세밀하게 나타내고 싶으면 xticks() 함수의 인자로 ticks 값을 나타내고 싶은 값을 설정하면 된다.



```python
x_value = np.arange(0, 100)
y_value = 2 * x_value

plt.plot(x_value, y_value, color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')

# X축의 눈금값을 np.arange(0, 100, 5)로 하여 0부터 99까지 5 스텝으로 20개를 표시하고
# 90도 회전한다
plt.xticks(ticks=np.arange(0, 100, 5), rotation=90)
```


*   범례는 여러 개의 그래프/차트를 한 Axes 내에 표시할 때 이들을 구분하려고 사용된다. 맷플롯립은 여러 개의 플롯을 하나의 Axes 내에서 그릴 수 있다. 범례를 표시하기 위해서는 plt.plot() 함수 내에 label 인자값을 설정해 해당 플롯에 명칭을 부여해 주어야 한다. 그런 다음 plt.legend()를 호출해 범례를 Axes 상에 나타낼 수 있다.
*   직접 Axes 객체 메서드를 통해서 시각화를 수행할 수도 있다. 먼저 plt.axes()를 이용해 Axes 개체를 추출한다. 그리고 plt.plot()은 Axes 객체의 plot()으로, plt.bar()은 Axes 개체의 bar()로 변경한다. 마찬가지로 plt.xlabel()과 plt.ylabel()은 각각 Axes 객체의 set_xlabel(), set_ylabel()로 변경하며, plt.title()은 Axes 객체의 set_title()로 변경한다. 하지만 plt.legend()는 Axes 객체의 set_legend()가 아닌 legend()로 적용되어야 한다.



```python
# X좌표값은 0~99까지 100개, 첫번째 Y좌표값은 X좌표값의 두 배, 두번째 좌표값은 X좌표값의 4배
x_value_01 = np.arange(0, 100)
y_value_01 = 2 * x_value_01
y_value_02 = 4 * x_value_01

# X좌표값과 첫번째 Y좌표값으로 녹색의 선그래프를 그린다. 이때 label은 'green line'으로 설정한다
plt.plot(x_value_01, y_value_01, color='green', label='green line')
# X좌표값과 두번째 Y좌표값으로 빨간색의 선그래프를 그린다. 이때 label은 'red line'으로 설정한다
plt.plot(x_value_01, y_value_02, color='red', label='red line')

plt.xlabel('x axis')
plt.ylabel('y axis')

# 개별 plot들의 label로 설정된 문자열을 범례로 표시한다
plt.legend()
plt.title("Hello plot")
plt.show()
```

```python
x_value_01 = np.arange(1, 10)
y_value_01 = 2 * x_value_01

# 마커를 포함한 빨간색 대시 선 그래프를 그리고, label은 'red line'으로 설정한다
plt.plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
# X값에 따른 Y값을 나타내는 초록색 막대 그래프를 그리고 label은 'bar plot'으로 설정한다
plt.bar(x_value_01, y_value_01, color='green', label='bar plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
# 두 개의 그래프에서 label 설정된 문자열 값을 범례로 표시한다
plt.legend()
plt.title('Hello plot')
plt.show()
```

```python
figure = plt.figure()
# Axes 객체를 추출하여 ax 변수에 할당한다
ax = plt.axes()

# plt.plot()은 ax.plot()으로, plt.bar()는 ax.bar()으로 변경한다
ax.plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax.bar(x_value_01, y_value_01, color='green', label='bar plot')

# plt.xlabel()은 ax.set_xlabel()로, plt.ylabel()은 ax.set_ylabel()로 변경한다
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

# ax.set_legend()가 아니라 ax.legend()이다
ax.legend()
# plt.title()을 plt.set_title()로 변경한다
ax.set_title('Hello plot')
plt.show()
```

### 여러 개의 subplot들을 이용해 개별 그래프들을 subplot별로 시각화하기



```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 3)) ## Axes 객체들을 배열 형태로 받아서 시각화한다.
```

반환된 Axes 객체들을 가지는 ax 변수는 튜플이 아니라 넘파이 ndarray 형태로 가지며, 각각 개별 원소인 ax[0]는 왼쪽에 생성된 Axes 객체를, ax[1]은 오른쪽에 생성된 Axes 객체를 가리킨다.

```python
print('ax type:', type(ax))
print('ax[0] type:', type(ax[0]))
print('ax[1] type:', type(ax[1]))
```



*   여러 개의 subplot들을 이용해서 시각화를 구현할 때에는 유의해야 할 사항이 잇다. 바로 개별 Axes를 직접 이용해 시각화를 구현해야 한다는 점이다. 이는 비단 선 그래프나 막대 그래프를 그리는 것에 국한되지 않고, 타이틀이나 축의 명칭, 눈금값, 범례를 설정할 때에도 개별 Axes에 직접 이들을 적용해야 한다는 것이다.
*   plt.subplots()의 인자인 nrows가 1이거나 또는 ncols가 1인 경우에는 1차원 배열 인덱싱 방식으로 Axes를 접근할 수 있다. (e.g. fig, ax = plt.subplots(nrows=1, ncols=3)이라면 ax[0], ax[1], ax[2]와 같은 방식으로 3개의 Axes 객체에 접근 가능하다) 하지만 nrows도 2 이상이고, ncols도 2 이상이 되면 1차원 배열 인덱싱 방식이 아니라 2차원 배열 인덱싱 방식으로 plot.subplots()로 반환된 Axes 객체를 접근해야 한다.



```python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

# figsize는 (6, 3)인 Figure과 2개의 Axes를 가지는 subplots를 반환한다
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))

# 개별 Axes 객체들을 각각 이용해 선 그래프와 막대 그래프를 그린다
ax[0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[1].plot(x_value_02, y_value_02, color='green', label='bar plot')
# 개별 Axes 객체 각각에 X축명을 설정한다
ax[0].set_xlabel('ax[0] x axis')
ax[1].set_xlabel('ax[1] x axis')
# 개별 Axes 객체 각각에 범례를 설정한다
ax[0].legend()
ax[1].legend()
# 개별 Axes 객체 각각에 타이틀을 설정한다
ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')
plt.show()
```

```python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

# figsize는 (6, 6)인 Figure와 2x2개의 Axes를 가지는 subplots를 반환한다
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))

# 2x2 Axes 객체들을 각각 이용해 선 그래프와 막대 그래프를 그린다
ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[0][1].bar(x_value_02, y_value_02, color='green', label='bar plot')
ax[1][0].plot(
    x_value_01, y_value_01, color='green', marker='o', linestyle='dashed', label='green line'
)
ax[1][1].bar(x_value_02, y_value_02, color='red', label='red bar')

# 개별 Axes 객체 각각에 X축명을 설정한다
ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')

# 개별 Axes 객체 각각에 범례를 설정한다
ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

plt.show()
```
